## Recursion

```python
 def fact(n):
     if n==0 or n==1:
         return 1
     elif n<0:
         print("negative fact does not exist")
     else:
         return n*fact(n-1)
```

```python
fact(2)
```

```python
fact(1)
```

```python
aa=[1,2,3,4]
def sum(aa):
    if len(aa)==0:
        return 1
    else:
        return aa[0]*sum(aa[1:])
```

```python
sum(aa)
```

```python
bb=[2,4,6,8]
def mul(bb):
    if len(bb)==0:
        return 1
    else:
        return bb[0]*mul(bb[1:])
        
```
