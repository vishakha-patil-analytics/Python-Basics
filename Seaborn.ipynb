## EDA (Exploratory data analysis)

#### 1.Scatterplot

```python
sns.scatterplot(data=df,x='Total_bill',y='tip')
```

```python
sns.scatterplot(data=df,x='Total_bill',y='tip',hue='day')
```

```python
sns.scatterplot(data=df,x='Total_bill',y='tip',hue='Gender')
```

#### 2.Countplot

```python
df['day'].unique()
```

```python
df['day'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
a=sns.countplot(data=df,x='Gender')
for i in a.containers:
    a.bar_label(i)
```

```python
x=sns.countplot(data=df,x='Gender',hue='time')
for i in x.containers:
    x.bar_label(i)
```

```python
aa=sns.countplot(data=df,x='Gender',hue='day')
for i in aa.containers:
    aa.bar_label(i)
```

#### 3.Violinplot

```python
sns.violinplot(data=df,x="tip",y="day",hue="Gender")
```

#### 4.Boxplot

```python
sns.boxplot(data=df,x="tip",y="Gender")
```

#### 5.Pairplot

```python
sns.pairplot(df)
```

#### 6.Catplot

```python
sns.catplot(data=df,x="Gender",y="tip")
```

#### 7.Displot

```python
sns.displot(df['Total_bill'],bins=13,kde=True)
```

#### 8.Jointplot

```python
sns.jointplot(data=df,x="Total_bill",y="tip",hue="Gender")
```
