### Merge

#### 1.Inner merge

```python
df1=pd.DataFrame({'country':['India','US','UK','UAE'],'Temp':[25,23,26,24]})
df1
```

```python
df2=pd.DataFrame({'country':['India','US','UK','Canada'],'Humidity':[80,83,86,84]})
df2
```

```python
inner_merge=pd.merge(df1,df2,how='inner',on='country')
inner_merge
```

```python
outer_merge=pd.merge(df1,df2,how='outer',on='country')
outer_merge
```

```python
left_merge=pd.merge(df1,df2,how='left',on='country')
left_merge
```

```python
right_merge=pd.merge(df1,df2,how='right',on='country')
right_merge
```
